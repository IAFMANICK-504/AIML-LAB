```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

df = pd.read_excel("weight of copper.xlsx")
df = df.drop("Unnamed: 0", axis=1)

# Clean newline characters
df = df.replace('\n', '', regex=True)
df['power_rating'] = pd.to_numeric(df['power_rating'], errors='coerce')
df['weight_of_coppe'] = pd.to_numeric(df['weight_of_coppe'], errors='coerce')
df = df.dropna()

K = {'low': 0, 'high': 1}
df['category'] = df['category'].map(K)

X = df[["power_rating", "weight_of_coppe"]]
Y = df[["category"]]

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)

clf = RandomForestClassifier(max_depth=2, random_state=0)
clf.fit(X_train, y_train.values.ravel())
```

```python
y_pred = clf.predict (X_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score
print("accuracy score",accuracy_score(y_test, y_pred))
print("mean_squared_error",mean_squared_error(y_test, y_pred))
from sklearn.metrics import confusion_matrix
```

```python
y_true = y_test
data = confusion_matrix(y_true, y_pred)
data
```
